```python
import pandas as pd
```

```python
PK_data=pd.read_csv('all_data.csv')
PK_data
```

```python
# null값만 있는 데이터프레임 생성
PK_occupied_na=PK_data.loc[PK_data.occupied.isnull()]
PK_occupied_na
```

```python
PK_filename=PK_occupied_na['filename'].str[:19]
PK_filename
```

```python
# null값을 가진 filename 요소 개수(중복값 찾기) 
cnt=PK_filename.value_counts()
cnt
```

```python
# 중복개수의 평균값
total=0
for i in range(len(cnt)):
    if cnt[i]!=1:
        total=total+cnt[i]
na_avg=total/len(cnt) 
print(na_avg)
```

```python
# 평균값보다 높은 값은 표에서 모두 제거, 그 외 null값들은 0으로 채우기
na_remove=[]
for i in range(len(cnt)):
    if cnt[i]>na_avg:
        na_remove.append(cnt.index[i])
```

```python
len(na_remove)
```

```python
PK_data_na_remove=pd.DataFrame()
```

```python
# 제거 리스트
for i in range(len(na_remove)):
    PK_data_na_remove=PK_data_na_remove.append(PK_data[PK_data['filename'].str[0:19]==na_remove[i]])
```

```python
# 제거 리스트 파일
PK_data_na_remove
```

```python
# 최종 파일에서 제거 리스트를 빼기
exfinal_PK_data=pd.merge(PK_data,PK_data_na_remove,how='outer',indicator=True)
exfinal_PK_data
```

```python
final_PK_data=exfinal_PK_data.query('_merge=="left_only"').drop(columns=['_merge'])
final_PK_data
```

```python
# 남은 null값 0으로 채우기
final_PK_data['occupied']=final_PK_data['occupied'].fillna(0)
```

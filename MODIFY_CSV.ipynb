```python
import pandas as pd
data_df = pd.read_csv('all_data_drop_na.csv')
```

```python
data_df
```

```python
len(data_df)
```

```python
print(data_df.iloc[0,8])
print(data_df.iloc[0,9])
print(data_df.iloc[0,10])
print(data_df.iloc[0,11])
```

```python
cnt = 0
for i in range(len(data_df)):
    data =  data_df.iloc[i]
    x_max = data.x_max.copy()
    x_min = data.x_min.copy()
    y_max = data.y_max.copy()
    y_min = data.y_min.copy()
    if data.x_max < data.x_min:
        data_df.iloc[i,8] = x_min
        data_df.iloc[i,9] = x_max
        cnt += 0.5
    if data.y_max < data.y_min:
        data_df.iloc[i,10] = y_min
        data_df.iloc[i,11] = y_max
        cnt += 0.5
print(cnt)
```

```python
data_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

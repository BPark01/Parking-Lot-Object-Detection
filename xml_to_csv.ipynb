```python
import os
import pandas as pd
```

```python
data_dir='C:/Users/owner/Desktop/bigdata/ProjectData (1)/ProjectData/PKLot/'
data_folder=os.listdir(data_dir)
data_xml=[]
for i in data_folder:
    data_xml_list=os.listdir(data_dir+i+'/xml')
    data_xml.append(data_xml_list)
    print(data_xml)
```

```python
data_dir+data_folder[0]+'/xml/'+data_xml[0][0]
```

```python
import xml.etree.ElementTree as ET
l_count=0
for l in data_folder:
    for i in data_xml[l_count]:
        tree=ET.parse(data_dir+l+'/xml/'+i)
        root=tree.getroot()
        #ID Occ  
        placeID=[]
        placeOcc=[]
        for child in root:
            try:
                placeID.append(child.attrib['id'])
                placeOcc.append(child.attrib['occupied'])
            except:
                placeOcc.append('NA')

        #x, y 4  
        x=[]
        y=[]
        xx=[]
        yy=[]
        count=1
        for neighbor in root.iter('point'):
            x.append(neighbor.attrib['x'])
            y.append(neighbor.attrib['y'])
            count+=1
            if count==5:
                count=1
                xx.append(x)
                yy.append(y)
                x=[]
                y=[]

        # 
        for j in range(0,len(xx)):
            xx[j]=list(map(int,xx[j]))
            yy[j]=list(map(int,yy[j]))

        #xmax, xmin, ymax, ymin 
        data_xmax=[]
        data_xmin=[]
        data_ymax=[]
        data_ymin=[]
        data_imageID=[]
        for k in range(0,len(xx)):
            data_xmax.append(max(xx[k]))
            data_xmin.append(min(xx[k]))
            data_ymax.append(max(yy[k]))
            data_ymin.append(min(yy[k]))
            data_imageID.append(i[:-4])

        # 
        df=pd.DataFrame({'imageID':data_imageID,'placeID':placeID,'placeOcc':placeOcc,'xmax':data_xmax,'xmin':data_xmin,'ymax':data_ymax,'ymin':data_ymin})

        #df->csv (  )
        df.to_csv('./xml_folder/'+l+'/'+i[:-4]+'.csv')
    l_count+=1
```
